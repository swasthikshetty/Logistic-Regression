```python
df1 =  pd.read_csv('bank-full.csv', sep=';')
df1.head()
```

```python
#for better visualization, store the tabularized data in seperate file 
# df1.to_csv("bank-full-tabularized.csv",encoding="utf-8")
```

```python
df2=pd.DataFrame(df1)
df2.head()
```

```python
df2.drop(["age","job","marital","education","contact","day","month"], axis=1, inplace=True)
df2.head(3)
```

```python
# pd.get_dummies(df1.default, prefix = 'default')
# pd.get_dummies(df1.housing, prefix = 'housing')
# pd.get_dummies(df1.loan, prefix = 'Loan')
# pd.get_dummies(df1.poutcome, prefix = 'poutcome')

df2 = pd.get_dummies(data=df2, columns=["default","housing","loan","poutcome"])
df2['Y'] = df2.y.map({'no':0,'yes':1})
df2.drop(["y"],axis=1, inplace=True)
df2.head()
```

```python
#check for NaN or null values

df2.iloc[:,:].isnull().values.any()
```

```python
# df = df[['x','z','y','q','p']]
#Reorder columns

df2 = df2[['Y','balance','duration','campaign','pdays','previous','default_no','default_yes','housing_no','housing_yes','loan_no','loan_yes','poutcome_failure','poutcome_other','poutcome_success','poutcome_unknown']]
df2.head(2)
```

```python
from sklearn.linear_model import LogisticRegression

x=df2.iloc[:,1:]
y=df2.iloc[:,0]

classifier = LogisticRegression()
classifier.fit(x,y)
```

```python
# predicted value to our original data frame
y_pred = classifier.predict(x)
df1["yy"] = y_pred
df1['yy'] = df1.yy.map({0:'no',1:'yes'})
df1.head()
```

```python
#Probablity Value

y_prob = pd.DataFrame(classifier.predict_proba(x.iloc[:,:]))  
print(y_prob)
```

```python
# concat y_prob to our data frame

new_df = pd.concat([df1,y_prob], axis=1)
new_df.head()
```

```python
new_df.to_csv("Predcited Term deposit.csv")
```
